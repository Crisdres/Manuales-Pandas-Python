Como se puede evidenciar existen dos elementos principales en los gráficos el uno es la figura (figure), y el otro los ejes (axes).

* Se puede definar a la figura como el espacio en el cual se visualizara el gráfico, por medio de este objeto se puede definar el tamaño y la forma del gráfico.

* Por otra parte el eje, es la gráfica como tal, donde se podra visualizar la representación gráfica con la cual se resume información. Dentro del eje tambien se puede identificar a:
- Axis: Objetos parecidos a una recta numérica, estos se encargan de establecer los limites del gráfico como tambien, de los **TICKS**, que son las marcas en el eje, y los **TICKLABEL**, que son las etiquetas de las TICKS.

Por medio del siguiente comando, se puede generar una figura y un eje al mismo tiempo 

```python
fig, axs = plt.subplots()
```

Modificar el rango de los ejes por medio de los metodos *axes.Axes.set_xlim()y axes.Axes.set_ylim()*

```python
fig,axs = plt.subplots()
############################################
axs.set_xlim(right=15,left=-4)
axs.set_ylim(bottom=-4,top=15)
```

Se puede modificar los títulos por medio del método  *set_title()*

```python
fig, axs = plt.subplots()
axs.set_xlim(right=15,left=-4)
axs.set_ylim(bottom = -4 , top = 15)
############################################
axs.set_title("grafico de ejemplo")
axs.set_xlabel("eje x")
axs.set_ylabel("eje y")
```

## Utilizar el paradigma de programación orientada a objetos para generar gráficos

Como se puede evidenciar se declaran las variables fig y axs como objetos de clase y posteriomente se invoca sus métodos.

```python
fig, axs = plt.subplots()
axs.set_xlim(right=13,left=-4)
axs.set_ylim(bottom = -4 , top = 125)
############################################
axs.set_title("grafico de ejemplo")
axs.set_xlabel("eje x")
axs.set_ylabel("eje y")
############################################
axs.plot(x, x, label='Linea')  
axs.plot(x, x**2, label='Linea cuadrática')
axs.legend()
```

## Utilizar funciones para generar gráficos múltiples

Para generar gráficos eficientemente se debe generar funciones, además añadiendo 2 parametros a la función plt.subplots(#,#), se puede generar multiples gráficos.

```python
def grafi_1(eje,vector):
    global x
    eje.set_xlim(-10,10)
    eje.plot(x, vector, label='Lineal') 
    eje.plot(x, vector**2, label='Cuadrática') 
    eje.set_title("gráfico de ejemplo")
    eje.set_xlabel("eje x")
    eje.set_ylabel("eje y")
    eje.legend()
```

Para este ejemplo se, incrementa el argumento **sigsize**, que sirve para incrementar o disminuir el tamaño de todo el grafico. Ademas, con el parametro **dpi**, se puede incrementar la calidad de la grafica.

```python
fig, (ax,ax2) = plt.subplots(1, 2,figsize=(10,3),dpi=80)
grafi_1(ax,y)
grafi_1(ax2,y2)
```

Al trabajar con gráficos de puntos, es mucho mas recomendable, utilizar statterplot, dado que permite regular el color y el tamaño del maker.

```python
data = {'a': [1,2,3,4,5],
        'c': [100,300,600,900,1200]}
data['b'] = [1,2,3,4,5]
data['d'] = [100,300,600,900,1200]
fig,ax=plt.subplots()
ax.scatter( "a",'b', c='c', s='d', data=data)
```

La misma sintaxis de plot, se puede utilizar en varias graficás, es decir, pasarle primero el el vector con valores para el eje x y posteriormente el vector con valores para el eje y. Adicionalmente para este ejemplo se utiliza el metodo **tight_layout()**, del objeto fig, esto nos aseguro que los subgraficos tendran el espasiado correcto.

```python
a=[1,2,3,4,5]
b=[1000,900,600,300,1200]

fig,[[ax,ax2],[ax3,ax4],[ax5,ax6]] = plt.subplots(3,2,figsize=(7,7),dpi=100)
fig.tight_layout()
ax.plot(a,b)
ax2.bar(a,b)
ax3.barh(a,b)
ax4.scatter(a,b)
ax5.stem(a,b,use_line_collection=True)
ax6.fill_between(a,b)
```

## Texto en el grafico

Para insertar texto en el grafico, se puede utilizar comandos de latex

```python
data = {'a': [1,2,3,4,5],
        'c': [100,300,600,900,1200]}
data['b'] = [1,2,3,4,5]
data['d'] = [100,300,600,900,1200]
fig,ax=plt.subplots()
ax.scatter( "a",'b', c='c', s='d', data=data)
ax.text(1, 3, r'$\mu=100,\ \sigma=15$')
```

Ademas del metodo text, se cuenta con el metodo anotate, el cual resulta mucho mas util, dado a su funcionalidad.

Para utilizarlo, se debe iterar en el por medio de un bucle for, con las cordenadas de x y de y.
Posteriormente, se llama al metodo; el primer parametro corresponde a el texto que se quiere visualizar en el grafico, y el segundo parametro **xy**, se debe pasar en forma de tupla, las coordenadas.

```python
a=[1,2,3,4,5]
b=[1,2,3,4,5]
fig,ax=plt.subplots()
ax.scatter(a,b)
ax.axis([0,6,0,6])
for i,j in zip(a,b):
    ax.annotate(i, xy=(i-0.2, j-0.2))
```

Un uso mas extenso para el metodo anotate, es la capacidad de insertar figuras para señalar objetos, dentro del grafico, para esto, se incrementan dos nuevos parametros, xytext, donde al igual que xy, va la cordenada donde va a iniciar la linea y xy que en este caso es donde finaliza, adicionalmente con arrowprops, se regula el colory el tamaño.

```python
a=[1,2,3,4,5]
b=[1,2,3,4,5]
fig,ax=plt.subplots()
ax.scatter(a,b)
ax.axis([0,6,0,6])

for i,j in zip(a,b):
    ax.annotate(i, xy=(i-0.2, j-0.2))
###############################################################    
for i,j in zip(a,b):
    ax.annotate('$Y=X$', xy=(i, j), xytext=(5, j),
                 arrowprops=dict(facecolor="red", shrink=0.01,width=1,headwidth=3))

for i,j in zip(a,b):
    ax.annotate('$Y=X$', xy=(i, j), xytext=(i, 5),
                 arrowprops=dict(facecolor="blue", shrink=0.1,width=1,headwidth=3))
ax.axis("off")
```

El grafico de areas, es de gran utilidad para comparar como intersecan dos vectores de datos, sin embargo, no se puede utilizar demasiadas variables, dado que su visualización se vuelve compleja al incrementar los indicadores.

## Grafico de pastel

Los gráficos de pastel son de gran utilidad cuando se requiere analizar la frecuencia relativa de una variable.

```python
# vector de frecuencias absolutas
conteo_visitas = [300.9, 81.8, 75.9, 60.7, 58.2, 39.3, 37.7, 37.6, 37.5, 35.4]
# vector de etiquetas de las freuencias
paises = ['Francia', 'España', 'EEUU', 'China', 'Italia',
          'México', 'Reino Unido', 'Turquía', 'Alemania', 'Tailandia']
# los trozos del pastel que seran resaltados
explode = [0, 0.2, 0, 0, 0, 0.9, 0, 0, 0, 0]
```

```python
plt.style.use("seaborn-pastel")
plt.pie(conteo_visitas, labels=paises, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=200)
plt.show()
```

## Grafico de caja y vigotes

Por medio del diagrama de caja y vigote, se puede identificar los valores atípitos y tambien, tener un primera acercamiento sobre la distribución de los datos.

# Graficas de lineas con dos ejes


```python
plt.style.use( 'tableau-colorblind10')
fig, ax1 = plt.subplots()
###############EJE_1###############
ax1.plot(x, y,"black")
##dos lineas en el mismo eje##
ax1.set_xlabel('ejex')
ax1.set_ylabel('pato', color="red")
ax1.tick_params(axis='y', labelcolor="yellow")

###############EJE_2###############
ax2 = ax1.twinx()  
ax2.plot(x, y2,"black")
ax2.plot(x,[10*i/i for i in range(1,11)] )
ax2.set_xlabel('ejex')
ax2.set_ylabel('vaca', color="blue")
ax2.tick_params(axis='y', labelcolor="green")


fig.tight_layout()
plt.show()
```

```python


ax=df.plot(kind='line', x='x', y=['y2',"lin"], color=['black',"blue"])

ax2=df.plot(kind='line', x='x', y='y', secondary_y=True,color='black', ax=ax)

ax.set_ylabel('pato',color="red")
ax2.set_ylabel('vaca',color="blue")

ax.tick_params(axis='y', labelcolor="green")
ax2.tick_params(axis='y', labelcolor="yellow")

plt.tight_layout()
plt.show()
```
