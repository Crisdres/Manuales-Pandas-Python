# Manual de graficas con Python (Matplotlib)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
x=np.arange(1,11)
y=np.random.randint(20,size=10)
y2=np.random.randint(30,size=10)
```

### Estilos de ggplot

```python
styles = plt.style.available
styles
```

# Grafica lineal (plot)

```python
plt.plot(x,y,y2)
plt.show
```

# Grafica de linea vertial (stem)

```python
plt.stem(x,y,use_line_collection=True)
plt.show()
```

## Grafico de series comparadas (fill_beetwen)

```python
plt.fill_between(x,y,label="lala",alpha=0.2)
plt.fill_between(x,y2,label="lala",alpha=0.2)
plt.show()
```

## Grafico de pastel

```python
# vector de frecuencias absolutas
conteo_visitas = [300.9, 81.8, 75.9, 60.7, 58.2, 39.3, 37.7, 37.6, 37.5, 35.4]
# vector de etiquetas de las freuencias
paises = ['Francia', 'Espaa', 'EEUU', 'China', 'Italia',
          'Mxico', 'Reino Unido', 'Turqua', 'Alemania', 'Tailandia']
# los trozos del pastel que seran resaltados
explode = [0, 0.2, 0, 0, 0, 0.9, 0, 0, 0, 0]
```

```python
plt.style.use("seaborn-pastel")
plt.pie(conteo_visitas, labels=paises, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=200)
plt.show()
```

## Grafico de caja y vigotes

```python
plt.boxplot([y2, y],
            notch=False, patch_artist=False,
            capprops=dict(color="black"),
            medianprops=dict(color="red"),
            whiskerprops=dict(color="green"))
plt.show()
```

## Histograma

```python
y3=np.random.normal(20,size=100)
plt.hist(y3, bins=9, edgecolor='black')
plt.show()
```

## Grafico de barras 

```python
plt.bar(paises, conteo_visitas, edgecolor='black')
plt.xticks(rotation=60)
plt.show()
```

## Grafico de barras horizontales

```python
plt.barh(paises, conteo_visitas, edgecolor='black')
plt.show()
```

## Grafico de escaleras 

```python
plt.step(x,y)
plt.show()
```

## Graficos de dispercion

```python
plt.style.use( 'tableau-colorblind10')
plt.style.use('Solarize_Light2')
plt.scatter(x,y)
plt.scatter(x,y2)
plt.show()
```

## Metodos y atributos de los graficos

```python
plt.title("tiutlo del grafico")
plt.xticks([1, 2, 3]"cambiar estas etiquetas", ['Japn', 'Alemania', 'Argentina'] "por estas"
          ,rotation=30,"rotar los ejes")
plt.xlabel("Altura media (cm)")
plt.ylabel("Muestras")
```

# Graficas de lineas con dos ejes


```python
plt.style.use( 'tableau-colorblind10')
fig, ax1 = plt.subplots()
###############EJE_1###############
ax1.plot(x, y,"black")
##dos lineas en el mismo eje##
ax1.set_xlabel('ejex')
ax1.set_ylabel('pato', color="red")
ax1.tick_params(axis='y', labelcolor="yellow")

###############EJE_2###############
ax2 = ax1.twinx()  
ax2.plot(x, y2,"black")
ax2.plot(x,[10*i/i for i in range(1,11)] )
ax2.set_xlabel('ejex')
ax2.set_ylabel('vaca', color="blue")
ax2.tick_params(axis='y', labelcolor="green")


fig.tight_layout()
plt.show()
```

# Grafico de lineas con dos ejes, usando un DataFrame

```python
df=pd.DataFrame([x,y,y2,[10*i/i for i in range(1,11)]])
df=df.T
df=df.rename(columns={0:"x",1:"y",2:"y2",3:"lin"})
```

```python


ax=df.plot(kind='line', x='x', y=['y2',"lin"], color=['black',"blue"])

ax2=df.plot(kind='line', x='x', y='y', secondary_y=True,color='black', ax=ax)

ax.set_ylabel('pato',color="red")
ax2.set_ylabel('vaca',color="blue")

ax.tick_params(axis='y', labelcolor="green")
ax2.tick_params(axis='y', labelcolor="yellow")

plt.tight_layout()
plt.show()
```
